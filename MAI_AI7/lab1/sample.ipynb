Подготовим данные для обучения, выделим целевой столбец и уберем столбцы коррелирующие с целевой переменной. Кодируем категориальные признаки с помощью One-hot encoding

```python
import pandas as pd
from sklearn.model_selection import train_test_split
df = pd.read_csv("./diabetes_dataset.csv")
x = df.drop(columns=["diagnosed_diabetes","diabetes_risk_score", "diabetes_stage"])
y = df["diagnosed_diabetes"]
x = pd.get_dummies(x, drop_first=True) #One-hot encoding
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

Обучим нашу модель и посморим на результат классификации

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
y_pred = knn.predict(x_test)
acc = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
print(acc,precision,recall,f1)
print(cm)
```
